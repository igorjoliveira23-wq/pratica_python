```python
# Caixa eletrônico
saque = int(input("Quanto deseja sacar da sua conta hoje?  "))
valor = saque
contador_nota = 0
cedula = 100

while True:
  if valor >= cedula:
    valor -= cedula
    contador_nota += 1
  else:
    print(f"O total de {contador_nota} cedula de ${cedula}")
    if cedula == 100:
      cedula = 50
    elif cedula == 50:
      cedula = 10
    elif cedula == 10:
      cedula = 5
    elif cedula == 5:
      cedula = 1
    contador_nota = 0
    if valor == 0:
      break
    if valor < 0:
      break
```
