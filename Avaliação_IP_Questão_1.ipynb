```python
df_populacao = pd.read_csv("/content/populacao_imc.csv")
display(df_populacao)
display(df_populacao.info())
```

```python
#a)CALCULAR O IMC DE CADA INDIVÍDUO

#criando cópia do data frame
df_analise = df_populacao
#criar coluna de IMC e limitar a duas casas decimais
df_analise['IMC']= df_analise['peso_kg'] / (df_analise['altura_m']**2)
df_analise['IMC'] = df_analise['IMC'].round(2)

#b)atribuir a classificação correspondente considerando a faixa etária

#calcular a idade para poder classificar
from datetime import datetime# importar o tempo
df_analise['data_nascimento'] = pd.to_datetime(df_analise["data_nascimento"], format="%Y-%m-%d")#converter "data_nascimento" para "datetime"
df_analise['Idade'] = df_analise['atual'] = datetime.now().year - df_analise['data_nascimento'].dt.year#calculando idade por meio da subtração entre ano atual e ano de nacimento(dt.year = apenas "ano")
df_analise = df_analise.drop(columns=['atual']) #ocultar coluna "atual"

#classificar faixa etária

#criando uma função que retorna a faixa etária do individuo com base na sua idade
def classificar_idade(idade):
  if idade < 12:
    return 'criança'
  elif  12 <= idade < 17:
    return 'adolescente'
  elif 18 <= idade <= 59:
    return 'adulto'
  else:
    return 'idoso'

df_analise['faixa etária'] = df_analise['Idade'].apply(classificar_idade)# a classificação da faixa etária será cada valor da coluna "idade" aplicado à função

#classificar IMC

def classificar_imc(IMC):
  if IMC <= 18.5:
    return 'baixo peso'
  elif  18.5 < IMC <= 24.9:
    return 'peso normal'
  elif 25 <= IMC <= 29.9:
    return 'sobre peso'
  elif 30 <= IMC <= 34.9:
    return 'obesidade grau I'
  elif 35 <= IMC <= 39.9:
    return 'obesidade grau II'
  elif 40 <= IMC:
    return 'obesidade grau III'

df_analise['classificação'] = df_analise['IMC'].apply(classificar_imc)

print('Esse é o data frame tratado')
#Ordenar por idade
df_analise = df_analise.sort_values(by='data_nascimento')
display(df_analise)

#c) análise dos dados

print('(i)contagem e o percentual de indivíduos por classificação de IMC dentro de cada faixa etária')

# criando uma "combinação" entre duas colunas(faixa etária e classificação) e chamndo de quantidade
df_i = df_analise.groupby(["faixa etária", "classificação"]).size().reset_index(name= 'quantidade')
#com base na quantidade, calculamos o percentual... exemplo:"adolescente com baixo peso/ soma de todos 'adolescentes'"
df_i['percentual'] = (df_i['quantidade'] / df_i.groupby('faixa etária')['quantidade'].transform(sum)) * 100
#limite de duas casa decimais
df_i['percentual'] = df_i['percentual'].round(2)
display(df_i)

print('(ii) o IMC médio, mediano e o desvio-padrão por faixa etária e sexo')

#calcular a média, mediana e desvio do IMC por faixa etária e sexo
df_ii = df_analise.groupby(['faixa etária','sexo'])['IMC'].agg( media_IMC= 'mean', mediana_IMC='median' , desvio_IMC='std').round(2).reset_index()
display(df_ii)

print('uma tabela que destaque os grupos (faixa etária × sexo) com maior proporção de classificações de risco')

#classificar quais grupos são de risco
risco = ['baixo peso', 'sobre peso', 'obesidade grau I', 'obesidade grau II', 'obesidade grau III']
#verificar se cada valor da coluna classificação está na lista risco
df_iii = df_analise['risco'] = df_analise['classificação'].isin(risco)
#calcular média atrelado à faixa etária e sexo e atribuir o nome de "percentual de  risco". Ordenar de forma decrescente
df_iii = df_analise.groupby(['faixa etária','sexo'])['risco'].mean().reset_index(name='percentual de risco').sort_values(by='percentual de risco', ascending=False)
#limitar a duas casas decimais e multiplicar por 100
df_iii['percentual de risco'] = df_iii['percentual de risco'].round(2) * 100
display(df_iii)


```
