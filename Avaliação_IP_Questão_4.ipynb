```python
n= int(input('Digite um número inteiro maior ou igual a zero: '))#usuário deve inserir um número
t = []#lista vazia para inserir o número
teste = [0, 1, 2, 9, 17, 49, 100]#lista para testar os resultados
t.append(n)#adiciona o número digitado pelo usuário na lista t


def classificar(n):# define a função que recebe um número n para classificar
  if n in (0,1):# verifica se o número é 0 ou 1
    return 'INDEF', None# retorna 'indefinido', pois 0 e 1 não são primos nem compostos
  for i in range(2,n):# percorre todos os números de 2 até n-1
    if n % i == 0:# verifica se n é divisível por i
      return 'SAFE',i# se for divisível, não é primo → retorna 'SAFE' e o divisor encontrado
  else:
    return 'MINA', None# se nenhum divisor foi encontrado, o número é primo → retorna 'MINA'

def classificar_teste(f):#mesma utilidade da função anterior, porém aplicada na lista "teste"
  if f in (0,1):
    return 'INDEF', None
  for j in range(2,n):
    if f % j == 0:
      return 'SAFE',j
  else:
    return 'MINA', None


for n in t:# percorre cada valor armazenado na lista t
  chave, valor  = classificar(n)# chama a função passando n e recebe dois retornos: status (chave) e payload (valor)
  print(f'CELULA: {n}')# exibe o número atual sendo analisado
  print(f'STATUS = {chave}')# mostra a chave da célula
  print(f'PAYLOAD={valor}')# mostra o valor da célula

print('=================itens comparativos======================')
#exatamente a mesma aplicabilidade do código anterior, porém aplicado na lista "teste"
for f in teste:
  chave, valor  = classificar(f)
  print(f'CELULA: {f}')
  print(f'STATUS = {chave}')
  print(f'PAYLOAD={valor}')
```
